```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# Retrieve data (updated daily)

# limit=0 to retrieve all available data, otherwise use limit=x where x=data points
btc_sent_url = 'https://api.alternative.me/fng/?limit=0'
btc_sent_raw = json.loads(requests.get(btc_sent_url).text)
```

```python
# Create and clean DataFrame

# Normalise nested JSON
df_btc_sent = json_normalize(btc_sent_raw['data'])

# Deleting unused column
df_btc_sent.drop('time_until_update', axis=1, inplace=True)

# Convert timestamp to datetime
# add '.dt.normalize()' at the end to change all times to midnight
df_btc_sent['timestamp'] = pd.to_datetime(df_btc_sent['timestamp'], unit='s')

# Renaming columns
df_btc_sent.columns = ['date', 'btc_sent_value', 'btc_sent']
```

```python
# Creating dummy variables for easier analysis
df_btc_sent = df_btc_sent.join(pd.get_dummies(df_btc_sent['btc_sent']))
```
